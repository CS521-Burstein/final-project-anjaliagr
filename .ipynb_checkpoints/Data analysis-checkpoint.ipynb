```python
def middle_income_func(economy):
    if economy in mid_inc_countries:
        return 1
    else:
        return 0
```

```python
def low_income_func(economy):
    if economy in lo_inc_countries:
        return 1
    else:
        return 0
```

```python
def high_income_func(economy):
    if economy in hi_inc_countries:
        return 1
    else:
        return 0
```
